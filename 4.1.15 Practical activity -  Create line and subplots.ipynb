```python
# Import Matplotlib, Seaborn, Numpy and Pandas.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load the Excel data using pd.read_excel.
ott = pd.read_excel('ott.xlsx')

# View the DataFrame.
print(ott.shape)
print(ott.dtypes)
print(ott.columns)
ott.head()
```

```python
# Load the CSV data using pd.read_csv.
movies = pd.read_csv('movies.csv')

# View the DataFrame.
print(movies.shape)
print(movies.dtypes)
print(movies.columns)
movies.head()
```

```python
# Merge the two DataFrames.
movies_ott = pd.merge(movies, ott, how='left', on = 'ID')

# View the DataFrame.
print(movies_ott.shape)
print(movies_ott.dtypes)
print(movies_ott.columns)
movies_ott.head()
```

```python
# Create a simple lineplot.
sns.lineplot(x='Year', y='IMDb', data=movies_ott)
```

```python
# Create a simple lineplot without confidence interval.
sns.lineplot(x='Year', y='IMDb', data=movies_ott, ci=None)
```

```python
# Create lineplots with specification.
sns.lineplot(x = 'Year', y = 'IMDb',
             data=movies_ott[movies_ott['Age'].isin(['16+', '18+'])],
             hue ='Age')
```

```python
# Create lineplots with specification but without confidence interval.
sns.lineplot(x = 'Year', y = 'IMDb',
             data=movies_ott[movies_ott['Age'].isin(['16+', '18+'])],
             hue ='Age', ci=None)
```
