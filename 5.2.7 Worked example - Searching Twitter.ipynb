```python
# Identify worldwide trends.
trends_worldwide = twitter_api.trends.available()

# Identify the number of available trends.
print(len(trends_worldwide))

# Example of trends_worldwide.
trends_worldwide[0]
```

```python
# Write a Twitter request.
list_of_names = [_['name'] for _ in trends_worldwide]

# List of first 10 cities
list_of_names[0:10]
```

```python
# Find Nairobi.
our_city = 'Nairobi'

# Create a variable.
list_of_names_our_city = [_ for _ in trends_worldwide if _['name'] == our_city]

# View the output.
print(len(list_of_names_our_city))

# Use index to find Nairobi.
list_of_names_our_city[0]
```

```python
# List of 'where on earth identifiers' (woeid)
list_of_names_our_city[0]['woeid']
```

```python
# Look at common trends in Nairobi.
nairobi_trends = twitter_api.trends.place(_id = list_of_names_our_city[0]['woeid'])

# View the output.
nairobi_trends
```

```python
# View output as a DataFrame.
# Import Pandas.
import pandas as pd

# Create a DataFrame.
nairobi_trends_pd = pd.DataFrame(nairobi_trends[0]['trends'])

# View the DataFrame.
nairobi_trends_pd
```

```python
# Narrow the list down to 100,000 tweets.
nairobi_trends_over100k_pd = nairobi_trends_pd[nairobi_trends_pd['tweet_volume'] > 100000]\
.sort_values('tweet_volume', ascending=False)

# View the output.
print(nairobi_trends_over100k_pd.shape)
nairobi_trends_over100k_pd
```

```python
# Save output as CSV file.
nairobi_trends_over100k_pd.to_csv('melbourne_over100k.csv', index=False)
```

```python
# Search for a common trend(s).
q = '#Arsenal'

# Set count to 100.
Count = 100
```

```python
# Read some tweets.
search_results = twitter_api.search.tweets(q=q, count=100)

statuses = search_results['statuses']
```

```python
for _ in range(5):
    print("Length of statuses", len(statuses))
    try:
        next_results = search_results['search_metadata']['next_results']
    except KeyError: # No more results when next_results doesn't exist
        break
        
    # Create a dictionary from next_results
    kwargs = dict([ kv.split('=') for kv in next_results[1:].split("&") ])
    
    search_results = twitter_api.search.tweets(**kwargs)
    statuses += search_results['statuses']

print(json.dumps(statuses[1], indent=1))
```
