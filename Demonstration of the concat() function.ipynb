```python
# A Simpler Concat() Solution. 

# Import Pandas. 
import pandas as pd
```

```python
# Read the 4 CSV files.
classlist_a = pd.read_csv('classlist_a.csv')
classlist_b = pd.read_csv('classlist_b.csv')
classlist_c = pd.read_csv('classlist_c.csv')
classlist_d = pd.read_csv('classlist_d.csv')
```

```python
#Concat the 4 files. 
classlist_final_2 = pd.concat([classlist_a, classlist_b, classlist_c, classlist_d])
```

```python
# View the DataFrame.
print(classlist_final_2.shape)
classlist_final_2
```

```python
# Sort Classlist_final_2 by Surname.
sorted_classlist_1 = classlist_final_2.sort_values(by = 'Surname')

# View the DataFrame.
sorted_classlist_1
```

```python
# Sort Classlist_final_2 by Student number. 
sorted_classlist_2 = classlist_final_2.sort_values(by = 'Student number')

# View the DataFrame.
sorted_classlist_2
```
