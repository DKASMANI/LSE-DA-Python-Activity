```python
# Import package.
import pandas as pd

# Load the csv data using pd.read_csv.
products = pd.read_csv('products.csv')

# View the shape of DataFrame.
print(products.shape)

# View the DataFrame.
products.head()
```

```python
# Products containing the word "fire". 

# Create a user-defined function
def contains_fire(a):
    """ does the product contain fire? """
    b = a.lower()
    return "fire" in b

# View the output.
print(contains_fire(a="Fire polished glass bracelet red"))
print(contains_fire(a="fire polished glass bracelet red"))
print(contains_fire(a="Enamel fire bucket cream"))
print(contains_fire(a="Enamel flower jug cream"))
```

```python
# Use the apply() function.
fire_products = products['Description'].apply(contains_fire)

# Filter the DataFrame.
products[fire_products]
```

```python
# Products containing the word "candle". 

# Create a user-defined function
def contains_candle(c):
    """ does the product contain candle? """
    d = c.lower()
    return "candle" in d

# View the output.
print(contains_candle(c="Fluted antique candle holder"))
print(contains_candle(c="Fluted antique Candle holder"))
print(contains_candle(c="Candleholder pink hanging heart"))
print(contains_candle(c="Ivory embroided quilt"))
```

```python
# Use the apply() function.
candle_products = products['Description'].apply(contains_candle)

# Filter the DataFrame.
products[candle_products]
```

```python
# Products containing the word "matches". 

# Create a user-defined function
def contains_matches(m):
    """ does the product contain matches? """
    n = m.lower()
    return "matches" in n

# View the output.
print(contains_matches(m="Retrospot small tube matches"))
print(contains_matches(m="Retrospot small tube Matches"))
print(contains_matches(m="Vintage union jack apron"))
```
